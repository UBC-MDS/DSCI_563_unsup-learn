```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# Generate points on the x-axis
x = np.linspace(-5, 5, 1000)

# Calculate the PDF of a Gaussian distribution
gaussian_pdf = norm.pdf(x, 0, 1)

# Calculate the PDF of a t-distribution with low degrees of freedom for heavy tails
t_pdf = t.pdf(x, df=1)  # df=1 for heavy tails

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, gaussian_pdf, label='Gaussian Distribution', linestyle='--')
plt.plot(x, t_pdf, label='t-Distribution (df=1)', linestyle='-')
plt.title('Comparison of Gaussian Distribution vs. Heavy-tailed t-Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
